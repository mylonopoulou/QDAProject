```python
#librerie

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.sandbox.stats.runs import runstest_1samp
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.stattools import acf
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
# Specifica la cartella contenente le immagini
folder_path = "/Users/emylonopoulou/Desktop/ProgettoPier/CorrectPieces"

# Crea una lista per le immagini
image_list = []

# Scansiona la cartella per i file
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    # Controlla se il file Ã¨ un'immagine
    if os.path.isfile(file_path) and filename.lower().endswith(('.bmp', '.jpg', '.jpeg', '.png')):
        try:
            # Apri l'immagine e aggiungila alla lista
            image = Image.open(file_path)
            image_list.append(image)
        except:
            # Ignora i file che non possono essere aperti come immagini
            pass

# Stampa il numero di immagini trovate
print("Numero di immagini trovate:", len(image_list))
```

```python
cv2_images = []
for image in image_list:
    # Conversione dell'immagine in un array NumPy
    np_image = np.array(image)

    # Conversione dell'immagine in un formato compatibile con OpenCV
    cv2_image = cv2.cvtColor(np_image, cv2.COLOR_RGB2BGR)

    # Aggiunta dell'immagine alla lista cv2_images
    cv2_images.append(cv2_image)
```

```python
m = 0
fixed_arr = []

for foto in cv2_images:
    #img = cv2_images[foto]
    m +=1
    #crop image 
    crop = foto[150:900, 20:980] 
    #rescale the image 
    newx,newy = foto.shape[1],foto.shape[0] 
    rescale = cv2.resize(crop,(newx,newy) ) 

    # converte l'immagine in scala di grigi 
    gray = cv2.cvtColor(rescale, cv2.COLOR_BGR2GRAY) 

    # applica un filtro di Gauss per ridurre il rumore dell'immagine 
    gray = cv2.GaussianBlur(gray, (5, 5), 0) 


    # applica la sogliatura di Otsu per ottenere un'immagine binaria 
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU) 

    # cerca i contorni nell'immagine 
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE) 

    # disegna i contorni dei pezzi individuati 
    img_contorni = rescale.copy() 
    l= 0 
    pezzi = [] 

    for c in contours: 
    # determina il rettangolo di delimitazione del contorno 
        x,y,w,h = cv2.boundingRect(c) 
        
        #selezione solo di cluster significativi 
        if w>200 and h>200 : 

            a= w 
            b= h 
            l+= 1 
        
            #disegno dei rettangoli 
            cv2.rectangle(img_contorni, (x,y), (x+a,y+b), (0,255,0), 2) 
        
            # calcolo di perimetro e area del contorno verde 
            area = a*b 
            perimetro = (a+b)*2 

            #numero di pixel bianchi 
            prova = thresh[y:y+b, x:x+a] 
            binary_img = np.array(prova, dtype=bool) 
            nwp= np.count_nonzero(binary_img)

            #analisi della struttura 
            white_pixels = np.where(prova== 255) 
            x_min, x_max = np.min(white_pixels[1]), np.max(white_pixels[1]) #white_pixels mi dava in output una tupla di due array: il primo riferito alla x, il secondo alla y
            y_min, y_max = np.min(white_pixels[0]), np.max(white_pixels[0]) 
            trimmed_img = prova[y_min:y_max, x_min:x_max] 

            #output = cv2.resize(trimmed_img, (256, 256), interpolation = cv2.INTER_CUBIC) 
            #plt.imshow(output)
            plt.imshow(trimmed_img, cmap="gray")
            plt.show()
            # Trova i contorni, il piu grande e del pezzo intero, gli altri dei buchi 
            #contorni, _ = cv2.findContours(output, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 

            contorni, _ = cv2.findContours(trimmed_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
            # Tieni solo quelli con area maggiore di una threshold per non considerare possibili puntini neri 
            contorni = [cnt for cnt in contorni if cv2.contourArea(cnt) > 10] #ora che sto considerando un pezzo, trovo tutti i contorni: uno sarÃ  quello del pezzo e gli altri quelli dei buchi
            areas = np.array([cv2.contourArea(cnt) for cnt in contorni]) #calcolo l'area per ogni contorno individuato e la metto nell'array areas
            perimeters = np.array([cv2.arcLength(cnt, True) for cnt in contorni]) 
            piece_index = np.argmax(areas) 
            holes_areas = np.delete(areas, piece_index) #dall'array con tutte le aree rimuovo l'area piÃ¹ grande, ovvero quella del pezzo, cosÃ¬ rimane l'elenco delle aree dei buchi
            holes_perimeters = np.delete(perimeters, piece_index) #stesso con i perimetri
            num_holes = len(holes_areas) 
            piece_area = areas[piece_index] - holes_areas.sum() 
            piece_perimeter = perimeters[piece_index] 
            Areamassima = np.max(areas)
            Areabuchi = holes_areas.sum()
            rpb = nwp/area

            #creo una lista con le informazioni dei pezzi nell'immagine 
            pezzo = { 
                "Foto" : m,
                "Piece n." : l, 
                "x": x, 
                "y": y, 
                "Length" : a, 
                "Height" : b, 
                "C. Perimeter" : perimetro, 
                "C. Area" : area, 
                "Numero pixel bianchi" : nwp, 
                "Number of holes" : num_holes,
                "Area of the holes" : Areabuchi,
                "Area in the contours" : Areamassima,
                "Piece surface": piece_area,
                "Piece perimeter": piece_perimeter, 
                "Rapporto pixel bianchi":rpb 
            } 

            fixed_arr.append(pezzo) 
            print(" - ", pezzo) 
            
print(fixed_arr)
```

```python
row = len(fixed_arr)
column = len(fixed_arr[0])

matrix = [[0]*column for i in range(row)]

for i in range(0,row,1):
    for j in range(0,column,1):
        item = fixed_arr[i]
        if j==0:
            matrix[i][j]=item["Foto"]
        if j==1:
            matrix[i][j]=item["Piece n."]
        if j==2:
            matrix[i][j]=item["x"]
        if j==3:
            matrix[i][j]=item["y"]
        if j==4:
            matrix[i][j]=item["Length"]
        if j==5:
            matrix[i][j]=item["Height"]
        if j==6:
            matrix[i][j]=item["C. Perimeter"]
        if j==7:
            matrix[i][j]=item["C. Area"]
        if j==8:
            matrix[i][j]=item["Numero pixel bianchi"]
        if j==9:
            matrix[i][j]=item["Number of holes"]
        if j==10:
            matrix[i][j]=item["Area of the holes"]
        if j==11:
            matrix[i][j]=item["Area in the contours"]
        if j==12:
            matrix[i][j]=item["Piece surface"]
        if j==13:
            matrix[i][j]=item["Piece perimeter"]
        if j==14:
            matrix[i][j]=item["Rapporto pixel bianchi"]


header = ['Foto', 'Piece number', 'x', 'y', 'Length', 'Height', 'C. Perimeter', 'C. Area', 'Numero pixel bianchi', 'Number of holes', 'Area of holes', 'Area in the contours', 'Piece surface', 'Piece perimeter', 'Rapporto pixel bianchi']

dataframe = pd.DataFrame(matrix)
dataframe.columns = header
print("Tabella dati: ")
print(dataframe)
```

```python

#plot parametri di controllo

print('Number of holes')
plt.plot(dataframe['Number of holes'], "o-")
plt.show()

print('Numero pixel bianchi')
plt.plot(dataframe['Numero pixel bianchi'], "o-")
plt.show()

print('Piece surface')
plt.plot(dataframe['Piece surface'], "o-")
plt.show()

print('Piece perimeter')
plt.plot(dataframe['Piece perimeter'], "o-")
plt.show()
```

```python
#check su foto difettose ed esclusione dei pezzi
data = dataframe.copy()

esclusi = data[(data['Number of holes']< 60) | (data['Numero pixel bianchi'] < 42000) | (data['Piece surface'] < 35000) | (data['Piece perimeter'] > 1300) ].index

data.drop(esclusi, inplace= True)
```

```python
#plot parametri di controllo dataset corretto

print('Number of holes')
plt.plot(data['Number of holes'], "o-")
plt.show()

print('Numero pixel bianchi')
plt.plot(data['Numero pixel bianchi'], "o-")
plt.show()

print('Piece surface')
plt.plot(data['Piece surface'], "o-")
plt.show()

print('Piece perimeter')
plt.plot(data['Piece perimeter'], "o-")
plt.show()
```

```python
stat_runs, pval_runs = runstest_1samp(data['Number of holes'], correction=False)

print('Runs test statistic ={:.3f}'.format(stat_runs))
print('Runs test pvalue ={:.3f}'.format(pval_runs))

if pval_runs<0.05:
    print('The null hypothesis is rejected: the process is not random')
else:
    print('The null hypothesis is accepted: the process is random')
```

```python
stat_shapiro, p_shapiro = stats.shapiro(data['Number of holes'])
print('Statistic = %.3f, p-val = %.3f' % (stat_shapiro, p_shapiro))
# interpret
alpha = 0.05
if p_shapiro > alpha:
    print('H0 is aeccepted: data are normal')
else:    
    print('Reject H0: data are not normal')


# Plot the qqplot
stats.probplot(data['Number of holes'], dist="norm", plot=plt)
plt.show()
```
